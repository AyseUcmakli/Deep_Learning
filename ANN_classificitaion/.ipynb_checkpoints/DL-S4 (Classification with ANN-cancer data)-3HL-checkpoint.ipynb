## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('Cancer', axis=1)
y = df['Cancer']
```

```python
seed = 42
#
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.10, random_state = seed)
#her seferinde aynı indekstekileri teste ayır
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Modelling & Model Performance

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(36, activation = "relu", input_dim = X_train.shape[1]))#dense katmanı ifade eder. o katmandaki nöron sayısı
model.add(Dense(18, activation = "relu"))
model.add(Dense(9, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))#binary classification yaptığımızda sigmoid kullanıyoruz. eğer multi olsaydı softmax kullanırdık.

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])#classificationda binaryde loss olarak bu fonk kullanılır..
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose=1)
#modeli 1000 defa modele ver. Her seferinde modele 32 case sunuyor .. step sayısı
```

```python
model.summary()
#1747 tane parametre var.. Non-trainable params:0 yani ögrenmeyen parametre yok..
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.head()
```

```python
loss_df.plot();
#400 den sonra overfittinge doğru gidiyor.. val_loss ve loss arası açılmaya başlıyor.. 
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
# buradaki değerler modelin son hali üzerinden sonuçları test datasında test ettiğimiz değerler.
```

```python
y_pred = model.predict(X_test) > 0.5
#y_pred = model.predict_classes(X_test) for tf 2.5.0
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### EarlyStopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(36, activation = "relu", input_dim = X_train.shape[1]))
model.add(Dense(18, activation = "relu"))
model.add(Dense(9, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
#The patience is often set somewhere between 10 and 100 (10 or 25 is more common), 
#but it really depends on your dataset and network.
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 15, restore_best_weights = True)
#patience :ne kadarlık bir epoch gidelim. burada 15 epoch boyunca loss da değişiklik yoksa duruyor..
#restore_best_weights = True olursa (Geriye doğru en iyi weightleri verir).15. epochu değil o zamana kadarki en iyi weighti  ver..
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
#önce 0.89u accuracy . şimdi iyileşme olduğunu görüyoruz..
```

```python
y_pred = model.predict(X_test) > 0.5
#y_pred = model.predict_classes(X_test) for tf 2.5.0
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
#önce hatamız 2+4 iken şmdi 1+3 ye düşmüş..iyi bir başarı elde edilmiş..
```

### Dropout

The Dropout layer randomly sets input units to 0 with a frequency of `rate`
at each step during training time, which helps prevent overfitting.

```python
from tensorflow.keras.layers import Dropout
#dropoutta amaç katmanlar arasında rastsal bir şekilde nöronların bazılarını kapatıp(rastsal) sonuç almak. overfittingi engellemeye çalışmak..
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(36, activation="relu", input_dim = X_train.shape[1]))
model.add(Dropout(0.5))

model.add(Dense(18, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(9, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(1, activation="sigmoid"))

opt = Adam(lr = 0.001)
model.compile(optimizer = opt, loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
#drop out u kullanmamız overfitti çözmüş.. 400 civarında overfite gitme durumu vardı ama şuan o kısımdan sonra ayrışma yok los ve val_coss arası açılmıyor..
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
#accuracy değerinde de artış sağlanmış..
```

```python
y_pred = model.predict(X_test) > 0.5
#y_pred = model.predict_classes(X_test) for tf 2.5.0
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Cross Validation

**Keras models** can be used in **scikit-learn** by wrapping them with the **KerasClassifier** or **KerasRegressor** class.

To use these wrappers you must define a function that creates and returns your Keras sequential model, then pass this function to the build_fn argument when constructing the KerasClassifier class.

The constructor for the KerasClassifier class can take default arguments that are passed on to the calls to model.fit(), such as the number of epochs and the batch size.

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_classifier():
    tf.random.set_seed(seed)
    classifier = Sequential()
    classifier.add(Dense(units = 36, activation = 'relu'))
    classifier.add(Dense(units = 18, activation = 'relu'))
    classifier.add(Dense(units = 9, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = "adam", loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
classifier_model = KerasClassifier(build_fn = build_classifier, batch_size = 32, epochs = 100, verbose=0)#verbose altta yapilan islemleri gosteriyordu 

scores = cross_validate(estimator = classifier_model, X = X_train, y = y_train,
                        scoring = ['accuracy', 'precision', 'recall', 'f1'], cv = 10)

df_scores = pd.DataFrame(scores, index = range(1, 11)).iloc[:, 2:]

df_scores
```

```python
df_scores_summary = pd.DataFrame({"score_mean" : df_scores.mean().values, "score_std" : df_scores.std().values},
                                 index = ["acc", "pre", "rec", "f1"])

df_scores_summary
```

## GridSearchCV

https://machinelearningmastery.com/grid-search-hyperparameters-deep-learning-models-python-keras/

```python
from tensorflow.keras.optimizers import Adam, Adadelta, RMSprop, Nadam
```

```python
def build_classifier(optimizer, learn_rate):
    tf.random.set_seed(seed)
    classifier = Sequential()
    classifier.add(Dense(units = 36, activation = 'relu'))
    classifier.add(Dense(units = 18, activation = 'relu'))
    classifier.add(Dense(units = 9, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer(learn_rate), loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 15, restore_best_weights = True)
```

```python
classifier_model = KerasClassifier(build_fn = build_classifier, validation_split = 0.1, epochs = 200)

parameters = {'batch_size': [32, 64],
              'optimizer': [Adam, RMSprop, Adadelta, Nadam],#
              'learn_rate': [0.001, 0.003, 0.005]}

grid_model = GridSearchCV(estimator = classifier_model,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs = -1,
                          verbose = 1)

grid_model.fit(X_train, y_train, callbacks = [early_stop])
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
#grid_model.cv_results_
```

```python
grid_result = pd.DataFrame(grid_model.cv_results_)[["mean_test_score", "std_test_score", "rank_test_score", "params"]].sort_values(by="rank_test_score")
grid_result
```

Your results may vary given the stochastic nature of the algorithm or evaluation procedure, or differences in numerical precision. Consider running the example a few times and compare the average outcome.

```python
y_pred = grid_model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
